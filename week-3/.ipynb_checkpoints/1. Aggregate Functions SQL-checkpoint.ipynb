```python
import pandas as pd
import sqlite3

yelp_data = pd.read_csv(yelp_dataset_url)
conn = sqlite3.connect('foobar.db')

yelp_data.to_sql("restaurants", conn, if_exists='append', index=False)
```

```python
type(yelp_data)
```

```python
type(conn)
```

Now we can select data from the restaurants table.

```python
cursor = conn.cursor()
# cursor.execute("select * from restaurants;")
# cursor.fetchall()
```

```python



sql = """
    WITH city_restaurants AS 
        (
            SELECT city, AVG(rating) as average_city_rating, count(*) as number_of_restaurants FROM RESTAURANTS GROUP BY city
        )
    SELECT city, number_of_restaurants, max(average_city_rating) from city_restaurants where number_of_restaurants > 20;
"""
sql
```

```python


cursor.execute(sql)
```

```python
cursor.fetchall()
```

And we can see the columns of the table, by executing the following: 

```python
cursor.execute('PRAGMA TABLE_INFO({})'.format('restaurants'))
cursor.fetchall()
```

```python
cursor.execute("SELECT category, AVG(rating) as avg_rating, COUNT(name) as count_restaurants FROM restaurants GROUP BY category HAVING avg_rating > 4 AND count_restaurants > 10;")
cursor.fetchall()
```
