```python
sql = '''
WITH restaurants_count AS (
    select city, AVG(rating) as avg_rating, COUNT(*) as number_of_restaurants from restaurants GROUP BY city
), subset_restaurants AS (
    SELECT * from restaurants_count where avg_rating > 4
) SELECT * from subset_restaurants where avg_rating > 4.2;
'''
```
