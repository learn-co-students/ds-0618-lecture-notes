### Adding categories

```python
import pandas as pd

poverty_df = pd.read_excel('./resources/Poverty.xls')
poverty_df.set_index('Country')
poverty_df['GNP'] = poverty_df['GNP'].replace("*", '0')
poverty_df.head()
```

```python
poverty_df['GNP'] = poverty_df.GNP.astype('int64')
```

```python
poverty_df.set_index('Country')
poverty_df.head()

poverty_df[poverty_df['Region'] == 1]
```

```python
poverty_df.Region = poverty_df['Region'].replace(6, 'Africa')
poverty_df.Region = poverty_df['Region'].replace(5, 'Asia')
poverty_df.Region = poverty_df['Region'].replace(4, 'MidEast')
poverty_df.Region = poverty_df['Region'].replace(3, 'AIDem')
poverty_df.Region = poverty_df['Region'].replace(2, 'LatAmerica')
poverty_df.Region = poverty_df['Region'].replace(1, 'EastEur')
```

```python
poverty_df
```

```python
poverty_df.Region.unique()
```

```python
poverty_df["Region"] = poverty_df["Region"].astype('category',categories=['EastEur', 'LatAmerica', 'AIDem', 'Asia', 'MidEast', 'Africa'])
```

```python
regions_df = pd.get_dummies(poverty_df.Region,prefix='region', drop_first=True)
```

```python
regions_df
```

```python
df = pd.concat([poverty_df,regions_df],axis=1)

# # now drop the original 'country' column (you don't need it anymore)
df.drop(['Region'],axis=1, inplace=True)
```
